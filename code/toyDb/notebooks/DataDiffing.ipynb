```python
import os, sys
sys.path.append(os.path.join(os.path.abspath(''), '../'))

from databases.ExperimentDb import (
    db,
    Environment,
    ImageOnlyExperiment,
    ImageOnlyResource,
    ImageOnlyShader,
    ImageOnlyTrace,
    init_from_default_db,
    packBytesToSpv
)

import logging
import json
import pickle
import peewee as pw
from tqdm import tqdm
import vkExecute

logger=logging.getLogger(__name__)

if not db.is_closed():
    db.close()

init_from_default_db()

IMPORT_SOURCE="4060-db-exported.dat"

with open(IMPORT_SOURCE, "rb") as fp:
    combinedDb = pickle.load(fp)

dataShaders = combinedDb["shaders"]
```

This code diffs selected shaders from export dataset to current dataset.

```python
candShaderIds = ['wtjBDD', 'lllBR7']

shaderIdToIdx = {elem["shader_id"]: idx for idx, elem in enumerate(dataShaders)}
remoteShdrSpv = {}
localShdrSpv = {}

for candShaderId in candShaderIds:
    # remote shader
    remoteShdrSpv[candShaderId] = packBytesToSpv(
        dataShaders[shaderIdToIdx[candShaderId]]["fragment_spv"]
    )

    # local shader
    localShdrSpv[candShaderId] = packBytesToSpv(
        ImageOnlyShader.get(
            ImageOnlyShader.shader_id == candShaderId
        ).fragment_spv
    )
```

```python
interested = 'lllBR7'    

spvProc = vkExecute.SpvProcessor()
spvProc.loadSpv(remoteShdrSpv[interested])

remoteFragDisText, fragErrMsgs = spvProc.disassemble()
print(remoteFragDisText)
```

```python
spvProc = vkExecute.SpvProcessor()
spvProc.loadSpv(localShdrSpv[interested])

localFragDisText, fragErrMsgs = spvProc.disassemble()
print(localFragDisText)
```
